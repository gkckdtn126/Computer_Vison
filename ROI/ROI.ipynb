```python
import selectivesearch
import cv2
import matplotlib.pyplot as plt
import os
%matplotlib inline

### 오드리헵번 이미지를 cv2로 로드하고 matplotlib으로 시각화 
img = cv2.imread('./data/audrey01.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print('img shape:', img.shape)

plt.figure(figsize=(8, 8))
plt.imshow(img_rgb)
plt.show()
```

```python
import selectivesearch 

#selectivesearch.selective_search()는 이미지의 Region Proposal정보를 반환 
_, regions = selectivesearch.selective_search(img_rgb, scale=100, min_size=2000) #selective_search:object가 있을만한 위치 찾아주는 모듈,scale: 크게 하면 큰 object위주로 알고리즘이 돌아간다, 
#작게하면 작은 object위주로 알고리즘이 돌아간다, min_size: object를 추천해주는데 2000사이즈 이상만 추천해달라

print(type(regions), len(regions))
```

```python
regions #첫번째labels: id가 붙는거임, 두번째꺼가 bounding box의 정보, size가 object의 사이즈, 두개 같이 나온건 비슷한 레이블끼리 합쳐야된다는 의미, 
```

```python
# rect정보만 출력해서 보기
cand_rects = [cand['rect'] for cand in regions]
print(cand_rects)
```

```python
# opencv의 rectangle()을 이용하여 시각화
# rectangle()은 이미지와 좌상단 좌표, 우하단 좌표, box컬러색, 두께등을 인자로 입력하면 원본 이미지에 box를 그려줌. 
#rect의 왼쪽 두개는 시작점, 오른쪽 두개는 너비와 높이

green_rgb = (125, 255, 51) #green 색으로 함
img_rgb_copy = img_rgb.copy() #원본 이미지를 copy 뜸
for rect in cand_rects:
    
    left = rect[0]
    top = rect[1]
    # rect[2], rect[3]은 너비와 높이이므로 우하단 좌표를 구하기 위해 좌상단 좌표에 각각을 더함. 
    right = left + rect[2]
    bottom = top + rect[3]
    
    img_rgb_copy = cv2.rectangle(img_rgb_copy, (left, top), (right, bottom), color=green_rgb, thickness=2) #그림 위에다가 사각형 그려주는 것, color는 색깔, thickness는 두께
    
plt.figure(figsize=(8, 8))
plt.imshow(img_rgb_copy)
plt.show()
```

```python
cand_rects = [cand['rect'] for cand in regions if cand['size'] > 10000]

green_rgb = (125, 255, 51)
img_rgb_copy = img_rgb.copy()
for rect in cand_rects:
    
    left = rect[0]
    top = rect[1]
    # rect[2], rect[3]은 너비와 높이이므로 우하단 좌표를 구하기 위해 좌상단 좌표에 각각을 더함. 
    right = left + rect[2]
    bottom = top + rect[3]
    
    img_rgb_copy = cv2.rectangle(img_rgb_copy, (left, top), (right, bottom), color=green_rgb, thickness=2)
    
plt.figure(figsize=(8, 8))
plt.imshow(img_rgb_copy)
plt.show()
```
